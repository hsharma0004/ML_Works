```python
import pandas as pd
import numpy  as np
```

Importing the Bengluru House Price dataset

```python
data = pd.read_csv('D:\\VS STUDIO\\GUI\\Streamlit\\Bengaluru_House_Data.csv')
```

```python
data.head(5)
```

```python
data
```

```python
data.drop(columns=['area_type','availability','society'],inplace=True)
```

```python
data
```

```python
data.dropna(inplace=True)
```

```python
data
```

```python
data.location.value_counts()
```

```python
# To remove the Leading or trailing whitespaces
data['location'] = data['location'].apply(lambda x:x.strip())
data.location.value_counts()
```

```python
location_stats = data.groupby('location')['location'].agg('count').sort_values(ascending=False)
location_stats
```

```python
location_entries_lessthan_10 = location_stats[location_stats <= 10]
location_entries_lessthan_10
```

```python
data['location'] = data['location'].apply(lambda x: 'others' if x in location_entries_lessthan_10 else x)
data['location'].value_counts()
```

```python
data['size'].unique()
```

```python
# Creating a new col bedrooms
# Give the 1st element of the list [0]
data['bedrooms'] = data['size'].apply(lambda x : int(x.split(' ')[0]))
data
```

```python
data.total_sqft.unique()
```

```python
# To eliminate the range we need to take its average
def clean(sqft):
  tokens = sqft.split('-')
  if len(tokens) == 2:
    return (float(tokens[0]) + float(tokens[1])) / 2
  else :
    try:
      return float(sqft)
    except:
      return None
```

```python
data['total_sqft'] = data.total_sqft.apply(clean)
```

```python
data['total_sqft'].unique()
```

```python
data.dropna(inplace=True)
```

```python
data.describe()
```

```python
data['sqft_per_bedroom'] = data['total_sqft'] / data['bedrooms']
```

```python
data.sqft_per_bedroom.describe()
```

```python
data2 = data[data['sqft_per_bedroom'] >= 300]
```

```python
data2
```

```python
data2['price_per_sqft'] = round(data2['price']*100000/data2['total_sqft'],2)
```

```python
data2.price_per_sqft.describe()
```

```python
data3 = data2[data2['price_per_sqft'] >= 2000]
```

```python
data3
```

```python
# Removed cols which are made to identify outliers and pre-process data 
data3.drop(columns=['size','sqft_per_bedroom','price_per_sqft'],axis=1,inplace=True)
```

```python
data3
```

```python
# To pre-process data we need to first make sure that the data is numeric and not be categorical
from sklearn.preprocessing import OneHotEncoder,StandardScaler
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.pipeline import make_pipeline
```

```python
from sklearn.compose import make_column_transformer
```

```python
col_trans = make_column_transformer((OneHotEncoder(sparse_output=False),['location']),remainder='passthrough')
```

```python
lr = LinearRegression()
Scalar = StandardScaler()
```

```python
model = make_pipeline(col_trans,Scalar,lr)
```

```python
data_input = data3.drop(columns=['price'])
data_output = data3['price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(data_input,data_output,test_size=0.2)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
input = pd.DataFrame([['Electronic City Phase II',2000.0,3.0,2.0,3]],columns=['location','total_sqft','bath','balcony','bedrooms'])
```

```python
model.predict(input)
```
